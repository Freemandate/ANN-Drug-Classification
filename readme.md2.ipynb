```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Display plots inline (only necessary in Jupyter Notebook)
%matplotlib inline
```

```python
# Create dataset
X = np.array([
    [5, 3],
    [10, 15],
    [24, 10],
    [30, 45],
    [85, 70],
    [71, 80],
    [60, 78],
    [55, 52],
    [80, 91]
])
```

```python
# Initial scatter plot of the dataset
plt.scatter(X[:, 0], X[:, 1], label='True Position')
plt.title("Scatter Plot of Dataset")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Set the number of clusters and fit KMeans model with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0)  # random_state for reproducibility
kmeans.fit(X)
```

```python
# Print cluster centers and labels
print("Cluster Centers:\n", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)
```

```python
# Scatter plot with clusters (2 clusters)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.title("Scatter Plot with 2 Clusters")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Change number of clusters to 3 and fit KMeans model
kmeans2 = KMeans(n_clusters=3, random_state=0)
kmeans2.fit(X)
```

```python
# Scatter plot with clusters (3 clusters)
plt.scatter(X[:, 0], X[:, 1], c=kmeans2.labels_, cmap='rainbow')
plt.title("Scatter Plot with 3 Clusters")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Scatter plot with clusters and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=kmeans2.labels_, cmap='rainbow', label='Clustered Data')
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], color='black', label='Centroids')
plt.title("Scatter Plot with 3 Clusters and Centroids")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
