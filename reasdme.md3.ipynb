```python
#Load Libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load KMeans Fuction
from sklearn.cluster import KMeans
```

```python
#Create Dataset
X= np.array([[5,3],
           [10,15],
           [24,10],
           [30,45],
           [85,70],
           [71,80],
           [60,78],
           [55,52],
           [80,91],])
```

```python
#Simple Scatter Plot of Dataset
plt.scatter(X[:,0],X[:,1], label='True Position')
```

```python
#Set number of clusters
kmeans = KMeans(n_clusters=2) 
```

```python
# Set the number of clusters and fit KMeans model with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0)  # random_state for reproducibility
kmeans.fit(X)
```
